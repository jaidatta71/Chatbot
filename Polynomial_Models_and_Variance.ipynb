```python
from google.colab import drive
drive.mount('/content/drive')
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
import plotly.express as px
```

```python
auto = pd.read_csv('drive/MyDrive/colab_activity8_1_starter/data/auto.csv')
#auto.describe()
X = auto.loc[:, ['horsepower']]
y = auto['mpg']

sample_auto = auto.sample(n=10, random_state=22)
print(type(sample_auto))
X_train = sample_auto.loc[:,['horsepower']]      ## double brackets to make it DF. Single bracket for float data type
print(type(X_train))
y_train = sample_auto['mpg']
#X_train
print(type(y_train))
y_train
```

```python
### GRADED
### YOUR SOLUTION HERE
model_predictions = {f'degree_{i}': None for i in range(1, 11)}

#print("Starting Dictionary of Predictions\n", model_predictions)
for i in range(1,11):

#create pipeline
  pipe = Pipeline([
     ('quad_features',PolynomialFeatures(degree=i, include_bias = False)),
     ('quad_model',LinearRegression())
  ])

  pipe.fit(X_train, y_train)
  preds = pipe.predict(X_train)
  model_predictions[f'degree_{i}'] = preds

# Answer check
model_predictions
```

```python
pred_df = pd.DataFrame(model_predictions)
pred_df
```

```python
pred_df.subtract(y, axis=0)
#pred_df.subtract(y, axis=0).head(12)
error_df = pred_df.subtract(y, axis=0)
```

```python
error_df.var()
#error_df.var().idxmax()    ..degree_3
error_df.var().idxmax()
abc = error_df.var().idxmax().split('_')[1]
print(f"\nThe degree with the highest variance in errors is degree {abc}.")
```
