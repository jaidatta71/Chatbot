#OH 5 - December 5, 2024

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
ibm = pd.read_csv("https://github.com/SavioSal/datasets/raw/master/HR-Employee-Attrition.csv")

ibm.head()
```

```python
ibm['Attrition'].value_counts().plot(kind='pie',explode=[0.1,0.1],autopct='%1.1f%%',colors=['g','r'])
print(ibm['Attrition'].value_counts())
```

```python
print(round(pd.crosstab(ibm['Attrition'],ibm['Gender'],normalize=True)*100,2))

plt.figure(figsize=(12,5))
plt.subplot(121)
plt.title('Gender Vs Attrition')
sns.countplot(ibm,x='Gender',hue='Attrition')

plt.subplot(122)
plt.title('Gender Vs MonthlyIncome')
sns.boxplot(x=ibm['Gender'],y=ibm['MonthlyIncome'],hue='Attrition',data=ibm)
```

```python
print(pd.crosstab(ibm['Attrition'],ibm['Department']))

plt.figure(figsize=(20,5))
plt.subplot(121)
plt.title('Department Vs Attrition')
sns.countplot(data=ibm,x=ibm['Department'],hue=ibm['Attrition'])

plt.subplot(122)
plt.title('Department Vs MonthlyIncome')
sns.boxplot(data=ibm,x=ibm['Department'],y=ibm['MonthlyIncome'],hue=ibm['Attrition'])
```

```python
pd.crosstab([ibm['Attrition']],ibm['JobRole'],normalize=True)*100

plt.figure(figsize=(25,14))
plt.subplot(211)
plt.title('JobRole Vs Attrition')
sns.countplot(data=ibm, x=ibm['JobRole'],hue=ibm['Attrition'])
plt.title('Attrition amongst different JobRoles',size=15)

plt.subplot(212)
plt.title('JobRole Vs MonthlyIncome')
sns.boxplot(data=ibm,x=ibm['JobRole'],y=ibm['MonthlyIncome'])
```

```python
# Boxplots to identify outliers for 'MonthlyIncome'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=ibm)
plt.title('Boxplot of Monthly Income by Attrition')
plt.show()
```

```python
ibm.groupby(by='JobRole')['MonthlyIncome'].mean().sort_values()

print(pd.crosstab(ibm['JobInvolvement'],ibm['Attrition']))

plt.figure(figsize=(22,16))
plt.subplot(221)
plt.title('JobInvolvement Vs Attrition')
sns.countplot(data=ibm, x=ibm['JobInvolvement'],hue=ibm['Attrition'])

plt.subplot(222)
plt.title('JobInvolvement Vs Monthly Income')
sns.boxplot(data=ibm, x=ibm['JobInvolvement'],y=ibm['MonthlyIncome'],hue=ibm['Attrition'])
```

```python
plt.figure(figsize=(22,16))
plt.subplot(221)
plt.title('Education Vs Attrition')
sns.countplot(data=ibm, x=ibm['Education'],hue=ibm['Attrition'])

plt.subplot(222)
plt.title('Education Vs MonthlyIncome')
sns.boxplot(data = ibm, x=ibm['Education'],y=ibm['MonthlyIncome'])
```

```python
print(pd.crosstab(columns=ibm['Attrition'],index=ibm['EducationField']))

plt.figure(figsize=(22,16))
plt.subplot(221)
plt.title('EducationField Vs Attriton')
sns.countplot(data=ibm,x=ibm['EducationField'],hue=ibm['Attrition'])

plt.subplot(222)
plt.title('EducationField Vs Monthly Income')
sns.boxplot(data=ibm,x=ibm['EducationField'],y=ibm['MonthlyIncome'])
```

```python
# 3. Distribution of Numerical Features
numerical_features = ['Age', 'DailyRate', 'MonthlyIncome', 'YearsAtCompany', 'TotalWorkingYears', 'YearsInCurrentRole', 'YearsSinceLastPromotion']
ibm[numerical_features].hist(bins=20, figsize=(14, 10), color='skyblue', edgecolor='black')
plt.suptitle('Distribution of Numerical Features', fontsize=16)
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(15,10))
plt.subplot(211)
plt.title('Distribution of Age',size=15)
sns.distplot(ibm['Age'])

plt.subplot(212)
plt.title('Age Wise Binning wrt Attrition',size=15)
sns.countplot(x=ibm['Age'],hue='Attrition',data=ibm)
```

```python
# plt.title('Distribution of Age',size=15)
sns.displot(data=ibm, x=ibm['Age'], hue='Attrition', kde=True)
```
