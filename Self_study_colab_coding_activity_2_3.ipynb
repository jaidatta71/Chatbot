### Problem 3: Histogram of Life Expectancy



Use `plt.hist()` to create a histogram of the `lifeExp` column.  Use the `plt.savefig()` method to save the image as `life_exp_hist.png` in the `images` folder.



````python
#Your plot here
pophist=plt.hist(gap_df['lifeExp'], alpha = 1, bins = 8, edgecolor = 'black', label = 'Life Expectency')
plt.savefig('/content/drive/MyDrive/colab_activity2_3_starter/images/life_exp_hist.png')
````

### Problem 4: Adding a title and human-readable label



Use the `plt.title()` method to add the title "Histogram of Life Expectancy".  Also, use the `plt.xlabel()` method to add a label to the x-axis "Life Expectancy".  Save your image as `hist2.png` to the `image` folder.



````python
#Your plot here
# alpha parameter controls the transparency of the histogram bars ()
# label is used for showing legend. use 'plt.legend()' method
pophist=plt.hist(gap_df['lifeExp'], alpha = 1, bins = 'auto', edgecolor = 'black', label = 'Life')
plt.title('Histogram of Life Expectancy')
plt.xlabel('Life Expectancy')
plt.legend()
plt.savefig('/content/drive/MyDrive/colab_activity2_3_starter/images/hist2.png')
````

### Problem 5: Histogram of GDP



Use matplotlib to create a histogram of the GDP data.  Add the title "Histogram of GDP per capita" and an x-label "GDP Per Capita".  Save your histogram as `gdp_hist.png` to the images folder.



````python
#Your plot here

pophist=plt.hist(gap_df['gdpPercap'], alpha = 1, bins = 'auto', edgecolor = 'black', label = 'GDP%')
plt.title('Histogram of GDP per capita')
plt.xlabel('GDP Per Capita')
plt.legend()
plt.savefig('/content/drive/MyDrive/colab_activity2_3_starter/images/gdp_hist.png')
````

### Problem 6: Customizing the Histogram




Set the `bins` argument to 50, the `edgecolor` to black, `alpha` to 0.5, `color` to light blue, and follow your histogram function with the `plt.grid()` function to add a grid to the plot.  Add the same title and x-label as problem 5.



````python
#Your plot here
pophist=plt.hist(gap_df['gdpPercap'], alpha = 0.5, bins = 50, edgecolor = 'black', label = 'GDP%', color = 'lightblue')
plt.grid()
plt.title('Histogram of GDP per capita')
plt.xlabel('GDP Per Capita')
plt.legend()
plt.savefig('/content/drive/MyDrive/colab_activity2_3_starter/images/gdp_hist.png')
````

### Problem 7: Scatterplot of GDP vs. Life Expectancy



`matplotlib` also has a scatterplot method that you can use with `plt.scatter()`.  Use this to create a scatterplot with the GDP column on the x-axis and Life Expectancy on the y-axis.  Add human-readable labels "GDP Per Capita" and "Life Expectancy (years)" respectively and a title "GDP vs. Life Expectancy".



````python
#Your plot here
plt.scatter( x=gap_df['gdpPercap'],  y=gap_df['lifeExp'], alpha = 0.5, edgecolor = 'black')
plt.xlabel('GDP Per Capita')
plt.ylabel('Life Expectancy (years)')
plt.grid()
plt.title('GDP vs. Life Expectancy')
````

[Back to top](#Index:)

### Problem 8: Logarithmic Transformation



Notice the non-linear relationship between GDP and Life Expectancy.  One common transformation is to take something that looks exponential and take a logarithm in order to transform the relationship into a linear one.  Draw a scatterplot like the above, but use `np.log()` to plot the logarithm of GDP vs. Life Expectancy.  Add appropriate x and y labels, and the title "Logarithm of GDP vs. Life Expectancy".



````python
#Your plot here
plt.scatter( x=np.log(gap_df['gdpPercap']),  y= gap_df['lifeExp'],  alpha = 0.5, edgecolor = 'black')
plt.xlabel('GDP Per Capita')
plt.ylabel('Life Expectancy (years)')
plt.grid()
plt.title('GDP vs. Life Expectancy')
````
